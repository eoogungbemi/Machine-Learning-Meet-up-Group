# Major Rules of using Matplotlib and  Seaborn

<b>Matplotlib Rules </b><br>
Understand the Figure and Axes: Matplotlib plots reside in a Figure object. Within this figure, one or more Axes objects can exist. Each Axes represents a plot, with its own coordinate system. Grasping this hierarchy is crucial for advanced customizations.


Label Your Axes and Graphs: Always label the x-axis, y-axis, and provide a title for your graph. This is essential for clarity and to make your plot self-explanatory.

Use Appropriate Scales: Depending on your data, consider changing the scale of your axes (linear, logarithmic, etc.) to better represent the relationships within your data.



<b>Seaborn Rules </b><br>
Choose the Right Plot Type: Seaborn simplifies the creation of many statistical plots. Selecting the correct plot type (e.g., scatter, line, histogram, boxplot) is crucial for accurately conveying the data’s story.

Utilize Seaborn’s Themes: Seaborn comes with preset themes and a high-level interface for controlling the appearance of Matplotlib figures. Utilize these to make your plots more aesthetically pleasing with minimal code.



<b>Matplotlib Parameters:</b>

This function accepts parameters that enable us to set axes scales and format the graphs. These parameters are mentioned below :-<br>

plot(x, y): plot x and y using default line style and color.<br>
plot.axis([xmin, xmax, ymin, ymax]): scales the x-axis and y-axis from minimum to maximum values<br>
plot.(x, y, color=’green’, marker=’o’, linestyle=’dashed’, linewidth=2, markersize=12):<br>
x and y co-ordinates are marked using circular markers of size 12 and green color line with — style of width 2<br>
plot.xlabel(‘X-axis’): names x-axis <br>
plot.ylabel(‘Y-axis’): names y-axis<br>
plot.title('ggggsgsggsb whehh')<br>

# Line Plot

A Line plot can be defined as a graph that displays data as points or check marks above a number line, showing the frequency of each value.<br>

Used for numeric data <br>
Used to show trends <br>
Compare two or more different variables over time<br>
Could be used to make predictions<br>

```python
# define data values
experience = [1,3,4,5,7,8,10,12]

salary = [6500, 9280, 12050, 13200, 16672, 21000, 23965, 29793]
 
plt.plot(experience, salary)  # Plot the chart
plt.title("Salary of Data Scientists by their experiences")
plt.show()  # display
```

```python
#Adding Labels to x and y

plt.plot(experience,salary)
plt.title("Salary of Data Scientists by their experiences")
plt.xlabel("Years of Experience")
plt.ylabel("Salary or Income")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 24, 36, 40, 55]

plt.plot(x, y, marker='*', linestyle='-', color='green', label='Data Trend')

plt.xlabel('X-axissssss')
plt.ylabel('Y-axisyyyyyy')
plt.title('Visualizing Data Trends with Line Plot')
plt.legend()
plt.grid(False)
plt.show()
```

# Bar plot

Bar Charts: Comparing Categorical Data<br>
A barplot (or barchart) is one of the most common types of graphic.<br>
It shows the relationship between a numeric and a categoric variable.<br>
Each entity of the categoric variable is represented as a bar. The size of the bar represents its numeric value.<br>
Compare data amongst different categories <br>
Ideal for more than 3 categories<br>
Can show large data changes over time<br>

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 24, 36, 40, 55]

plt.bar(x, y)

plt.xlabel('X-axissssss')
plt.ylabel('Y-axisyyyyyy')
plt.title('Visualizing Data Trends with Line Plot')
plt.legend()
plt.grid(False)
plt.show()
```

# Histograms

A histogram is a graph showing frequency distributions.<br>
It is a graph showing the number of observations within each given interval.<br>
Used for continuous data <br>
We use hist() function in order to create histograms <br>
Summarize large datasets graphically <br>
Compare multiple distributions<br>

```python
# Generate random data for the histogram
data = np.random.randn(1000)
 
# Plotting a basic histogram
plt.hist(data, bins=80, color='skyblue', edgecolor='black')
 
# Adding labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')
 
# Display the plot
plt.show()
```

# Classical  Example

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

### Loading data

```python
df = pd.read_csv("ToyotaCorolla.csv")
```

```python
df.sample(5)
```

```python
df.shape
```

# Line chart

```python
## Matplotlib
plt.figure(figsize=(12, 8))
plt.plot(df.index, df['Price'])
plt.xlabel('Index')
plt.ylabel('Price')
plt.title('Price by index')
plt.legend()
plt.show()
```

```python
## Seaborn

plt.figure(figsize=(8, 6))
sns.lineplot(x=df.index, y='Price', data=df)
plt.xlabel('Index')
plt.ylabel('Price')
plt.title('Line Chart - Price')
plt.show()

plt.show()
```

# Barchart

```python
# Matplotlib Bar Chart
plt.figure(figsize=(10, 6))
plt.bar(df['FuelType'], df['Price'], color='skyblue')
plt.xlabel('Fuel Type')
plt.ylabel('Price')
plt.title('Price by Fuel Type')
plt.show()
```

```python
# Seaborn Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(x='FuelType', y='Price', data=df) ### Key  difference
plt.xlabel('Fuel Type')
plt.ylabel('Price')
plt.title('Price by Fuel Type')
plt.show()
```

# Scatterplot

```python
#  Maptplotlib Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['KM'], df['Price'], color='green')
plt.xlabel('KM')
plt.ylabel('Price')
plt.title('Scatter Plot - KM vs Price')
plt.show()
```

```python
# Seaborn Scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='KM', y='Price', data=df)
plt.xlabel('KM')
plt.ylabel('Price')
plt.title('Scatter Plot - KM vs Price')
plt.show()
```

```python
# Matplotlib Scatter Plot  for 3 columns
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Price'], c=df['HP'], cmap='viridis')
plt.colorbar(label='Horsepower')
plt.xlabel('Age')
plt.ylabel('Price')
plt.title('Price vs Age by Horsepower')
plt.show()
```

```python
# Seaborn Scatter Plot  3  columns
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Price', data=df, hue='HP', palette='viridis')
plt.xlabel('Age')
plt.ylabel('Price')
plt.title('Price vs Age by Horsepower')
plt.show()
```

# Histogram

```python
plt.figure(figsize=(12, 6))
plt.hist(df['Price'], bins=10, edgecolor='red')
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Price Distribution')
plt.show()
```

```python
# Histogram
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Price',bins=10, kde=True)
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Price Distribution')
plt.show()
```
